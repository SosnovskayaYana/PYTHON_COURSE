```python
import pandas as pd


df = pd.read_csv('sample_data/california_housing_train.csv')
```

```python
import seaborn as sns
```


Изобразите отношение households к population с помощью точечного графика

```python
sns.scatterplot(data=df, x="households", y="population")
```

Визуализировать longitude по отношения к median_house_value, используя линейный график

```python
sns.relplot(data = df, x = 'longitude', y = 'median_house_value', kind = 'line')
```

```python
sns.relplot(data = df, x = 'longitude', y = 'median_house_value', kind = 'line', height = 10, aspect = 3)
```

Представить гистограмму по housing_median_age

```python
sns.histplot(data = df, x = 'housing_median_age')
```

Изобразить гистограмму по median_house_value с оттенком housing_median_age

```python
sns.displot(df, x="median_house_value", hue="housing_median_age")
```

```python
sns.displot(df, x="median_house_value", hue="housing_median_age", legend = False, bins = range(0, 500001, 50000))
```

```python
sns.histplot(data=df, x='median_house_value', hue='housing_median_age', legend=False, bins=range(0,500001, 50000), multiple='stack')
```

убираем нулевые значения

```python
penguins.dropna(inplace = True)
```

соотношение "flipper_length_mm" к "body_mass_g" 

```python
sns.scatterplot(penguins, x="flipper_length_mm", y="body_mass_g")
```

соотношение 'bill_length_mm' к 'bill_depth_mm' с оттенком 'species')

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
height - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
penguins.loc[penguins['bill_length_mm'] >= 42, 'height_group'] = 'high'
penguins.loc[(penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] < 42), 'height_group'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'height_group'] = 'low'
```

```python
penguins
```

гистограмма по 'flipper_length_mm', с оттенком'height_group'

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='height_group', multiple='stack')
```
